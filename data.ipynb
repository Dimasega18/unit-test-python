```python
import re
import string

nama = df_new['Provinsi/Kabupaten/Kota'].values
karakter_unik = string.punctuation
data = []
for x in nama :
    data.append(bool(re.search('[!"#$%&\'()*+,-\./:;<=>?@[\\]^_`{|}~]',x)))
    
nama_unik = df_new.loc[data]

kabupatenkota_unik = nama_unik.loc[~(nama_unik['Provinsi/Kabupaten/Kota'] == 'KEP. BANGKA BELITUNG' )]

provinsi_unik = nama_unik.loc[(nama_unik['Provinsi/Kabupaten/Kota'] == 'KEP. BANGKA BELITUNG' )]

```

```python
provinsi = [True if x.isupper() else False for x in df_new['Provinsi/Kabupaten/Kota']]
df_new.loc[provinsi]._append(provinsi_unik,ignore_index=True).drop_duplicates()
```
